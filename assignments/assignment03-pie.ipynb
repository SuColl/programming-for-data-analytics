## Import modules and data

```python
# Import modules
# Pandas for data manipulation and analysis
import pandas as pd
# Pyplot for plotting functions
import matplotlib.pyplot as plt
```

```python
# Read data at URL into dataframe
URL = "https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download"
df = pd.read_csv(URL)
```

```python
# Look at dataframe head and info to see what has been loaded:
df.head()
```

```python
df.info()
```

```python
df.Email.head()
```

## Analysis of the Email Domains

```python
# To add a column to the dataframe that contains just the email domains, use the str.split method (as per Andrew's lecture example.)
# Documentation at: https://pandas.pydata.org/docs/reference/api/pandas.Series.str.split.html

df['Domain'] = df.Email.str.split('@').str[1]

df.Domain.head()
```

```python
# Use value_counts() method to count the number of times each domain occurs 
# in the domain column.
# Documentation: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.value_counts.html
domain_count = df.Domain.value_counts()

# Look at the results
print(domain_count)
```

## Plotting

```python
# The index of the domain count series will provide the plot labels.
domain_count.index
```

```python
# Plot the pie chart. Worked example of Matplotlib's pie chart is available at 
# https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html

# Set plot title
plt.title('Distribution of Email Domains in Dataset')
# Set parameter to make the pie wedges explode
explode=(0.1, 0.1, 0.1)
# Set chart colours
colors=["#648FFF", "#DC267F", "#FFB000"]

# Call the pie chart plot, with percentage labels
plt.pie(domain_count, labels = domain_count.index, explode=explode, 
        autopct='%1.1f%%', pctdistance=1.15, labeldistance=.25,
        colors=colors)

plt.show()
```

Note that the email domains are very evenly distributed in this data.
